#### dealing with outliers

```python
def plot(data,column):
    fig,(ax1,ax2)=plt.subplots(2,1)
    sns.distplot(data[column],ax=ax1)
    sns.boxplot(data[column],ax=ax2,x=train_df[column])
```

```python

plot(train_df,'Strength')
```

```python

def treat_outliers(data,column):
    q1 = (np.percentile(data[column],25))
    q3 = (np.percentile(data[column],75))
    iqr = q3-q1
    minn = q1 - (1.5 * iqr)
    maxx = q3 + (1.5 * iqr)
    for i in range(len(data[column])):
        val = data[column].iloc[i]
        if val < minn:
            data[column].iloc[i] = minn
        if val > maxx:
            data[column].iloc[i] = maxx
            
           
```

```python
from sklearn.model_selection import RepeatedKFold


n_folds = 5
n_repeats = 3

X = train_df.drop(columns=["Strength"])
y = train_df.Strength
```

```python
test_df1 = test_df.drop("id", axis=1)
```

```python
from sklearn.ensemble import GradientBoostingRegressor


k_fold = RepeatedKFold(n_splits=n_folds, n_repeats=n_repeats, random_state=42)

modelsSGB = []
predsSGB = []

par =  {
    'n_estimators': 1000,
    'learning_rate': 0.0095,
    'min_samples_split': 4,
    'max_features': "sqrt",
    'validation_fraction': 0.2,
    'n_iter_no_change': 5,
    'tol': 0.01,
    'random_state': 0,    
}


for train_index, test_index in k_fold.split(X, y):
    
    model = GradientBoostingRegressor(**par)
    
    model.fit(X, y)
    modelsSGB.append(model)
    predsSGB.append(model.predict(test_df1))
```

```python
predSGR = np.average(np.array(predsSGB),axis=0)
```

```python
coef = 0.5

a  = coef
```

```python
subm = pd.DataFrame()
subm["id"] = test_df.index + 5407
subm = subm.set_index("id")
subm["Strength"] = predSGR * a 

subm.to_csv('submission.csv')
```
